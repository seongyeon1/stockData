## Scoring

- EPS = 기업의 당기순이익 / 총 발행 주식수
    - 주당순이익을 의미하며 EPS가 높을 수록 투자 가치가 높다고 할 수 있다
    - 현재 주가와 함께 비교해주어야만 의미가 있다. EPS 대비 현재 주식의 수준이 저렴할 경우에 저평가로써 투자매력이 생기게 되기 때문

- BPS = 주당순자산가치 (Book-value per share) : 주당 순자산가치는 기업의 순자산을 발행 주식수로 나눈 것

- PER(주가수익비율) = 주가 / EPS = 시가총액/주식수 / 순이익/주식수 = 시가총액 / 순이익
    - 주식 가치가 고평가됐는지 저평가됐는지를 알 수 있는 매우 중요한 지표
- PBR(주당순자산비율)  = 주가 / BPS = 시가총액/주식수 / 순자산/주식수 = 시가총액 / 순자산
    - 낮을수록 저평가되어 있다는 의미이므로 매수해야 한다.
    - 따라서 평균보다 낮은 주식을 1 점 평균이상의 주식에 0 점을 부여.
- ROE(자기자본수익률) : 당기순이익 / 자기자본 * 100
    - 최근 5 년간의 ROE 의 평균 혹은 최근 ROE 의 가중치를 높인 가중 평균을 지표로 활용할 예정

- 베타
    - 1 미만 : 안정적 투자, 큰 수익 기대하기 어려움
    - 1 : 시장과 함께 변동하는 주가
    - 1 초과 : 위험적 투자, 큰 수익/손실 가능
    - 0 미만 : 마이너스 베타는 시장과 반대로 가는 put 옵션, inverse ETF가 가지는 수치
