## 2. Data preprocessing

- 숫자 단위 통일 
- 분석 하기로 한 column 만 추출

```python

def _conv_to_float(s):
    if s[-1] == '%':
        s = s.replace('%', '')
    if s[-1] in list('BMK'): # covert dollar measurement
        powers = {'B': 10 ** 9, 'M': 10 ** 6, 'K': 10 ** 3, '': 1}
        m = re.search("([0-9\.]+)(M|B|K|)", s)
        """
        re.serach(condition, string)- find the condition in the string 
        condition 
        (pattern) - find pattern and save mathced data 
        [filter]+ = find 0-9 , expreess any character behind as escape , + ( find more than one character satisfying the filter condition)
        (|||) one of M, B, K can come 
        """
        if m:
            val, mag = m.group(1), m.group(2) #gorup(1) -  ([0-9\.]+), group(2) -(M|B|K|)
            return float(val) * powers[mag]
    try:
        result = float(s)
    except:
        result = None
    return result
```

```python
sales =_conv_to_float(sales)
revenue =_conv_to_float(revenue)
marcap =_conv_to_float(marcap)
per =_conv_to_float(per)
pbr =_conv_to_float(pbr)
roa =_conv_to_float(roa)
roe =_conv_to_float(roe)
volume =_conv_to_float(volume)
price =_conv_to_float(price)
beta  =_conv_to_float(beta)
dividend =_conv_to_float(dividend)
```

## 1 & 2 통합 function (크롤링 + 데이터 전처리 )


```python
#== crawling , data pre processing #
"""
combined above functions 
"""

def stock_factors(sym): # arg. ticker
    headers = {'User-Agent': 'Mozilla/5.0'}
    r = requests.get('https://finviz.com/quote.ashx?t='+sym, headers=headers)

    soup = BeautifulSoup(r.text)
    tables = soup.find_all('table')

    snapshot_table2 = soup.find('table', attrs={'class': 'snapshot-table2'})
    tables = pd.read_html(str(snapshot_table2))
    df = tables[0]
    df.columns = ['key', 'value'] * 6

    ## convert columns to rows
    df_list = [df.iloc[:, i*2: i*2+2] for i in range(6)]
    df_factor = pd.concat(df_list, ignore_index=True)
    df_factor.set_index('key', inplace=True)

    #apply to the columns
    v = df_factor.value
    sales = _conv_to_float(v['Sales'])
    revenue = _conv_to_float(v['Income'])
    marcap = _conv_to_float(v['Market Cap'])
    per = _conv_to_float(v['P/E'])
    pbr = _conv_to_float(v['P/B'])
    roa = _conv_to_float(v['ROA'])
    roe = _conv_to_float(v['ROE'])
    volume = _conv_to_float(v['Volume'])
    price = _conv_to_float(v['Price'])
    beta = _conv_to_float(v['Beta'])
    dividend = _conv_to_float(v['Dividend %'])


    return {"Sales":sales, "Revenue":revenue, "MarCap":marcap, "PER":per, "PBR":pbr, "ROA":roa, "ROE":roe, "Volume":volume, "Price":price, "Beta":beta, "DIVIDEND":dividend}
```

```python
def json_maker(stocklist):

    for ix, row in stocklist.iterrows(): 
        sym, name = row['Symbol'] , row['Name'] #ticker, company name 
        json_fn = folder + f'{sym}.json'
        if os.path.exists(json_fn):
            print('skip', json_fn)
            continue

        try: 
            factors = stock_factors(sym)

        except ValueError:
            continue  
        except ImportError:
            continue

        with open(json_fn, 'w') as f:
            json.dump(factors, f)
        print(sym, name)
    
```

```python
def df_maker(stocklist):
    tmp_list = stocklist.copy()
    for ix, row in tmp_list.iterrows():
        sym, name = row['Symbol'], row['Name']
        json_fn = folder + f'{sym}.json'


    # open json files saved in json maker function
    # if there's an error in opening json file, save as none
        try: 
            with open(json_fn, 'r') as f:
                factors = json.load(f)

    #make another df with fundamental data
            for col in ["Sales","Revenue","MarCap", "PER", "PBR", "ROA", "ROE", "Volume", "Price", "Beta", "DIVIDEND"]:
                tmp_list.loc[ix,col] = factors[col]
        except:
            for col in ["Sales","Revenue","MarCap", "PER", "PBR", "ROA", "ROE", "Volume", "Price", "Beta", "DIVIDEND"]:
                tmp_list.loc[ix,col] = None
            print("error for", sym, name)
            
            
    return tmp_list
    
```

### 점수화 - SCORING FUNCTION DEF. SCORING, 전체 s&p 500

```python
def scoring (p_attribute_list, n_attribute_list, df, standard_df): #postive - the bigger the better attribute list, negative = the lower the better attribute list 
  for idx, row in df.iterrows():
    for att in p_attribute_list:
      if row[att] >= standard_df.loc[row["Sector"],att] : 
        df.loc[idx,att+"_score"]=1
      else:
        df.loc[idx,att+"_score"]=0
    for att in n_attribute_list:
      if row[att] <= standard_df.loc[row["Sector"],att] : 
        df.loc[idx,att+"_score"]=1
      else:
        df.loc[idx,att+"_score"]=0
    df.loc[idx, "total_score"] = df.loc[idx, df.columns.str.contains("_score")].sum()
  return df
```

```python
p_check_list = ['Sales','Revenue','MarCap','ROA','ROE']
n_check_list = ['PER']
```

```python
try_df = scoring(p_check_list, n_check_list, tmp, standard)
```
