```python
import warnings
warnings.filterwarnings('ignore')

from pykrx import stock
import pandas as pd

import requests
import datetime
```

```python
def get_sector(dt):
    '''
    각 종목별로 그날 그날의 데이터를 가져온다
    '''
    kospi = pd.read_csv('./data/kospi.csv', encoding='cp949')
    kosdaq = pd.read_csv('./data/kosdaq.csv', encoding='cp949')

    kospi = kospi[['종목코드','종목명','시장구분','업종명']]
    kosdaq = kosdaq[['종목코드','종목명','시장구분','업종명']]

    stock_list = pd.concat([kospi, kosdaq])

    #stock_fud = pd.DataFrame(stock.get_market_fundamental('20200102','20220725','005930'))
    stock_fud = pd.DataFrame(stock.get_market_fundamental_by_ticker(date=dt, market="ALL"))
    stock_fud = stock_fud.reset_index()
    stock_fud.rename(columns={'티커':'종목코드'}, inplace=True)

    stock_data = pd.merge(stock_list, stock_fud, left_on='종목코드', right_on='종목코드', how='outer')

    stock_price = stock.get_market_ohlcv_by_ticker(date=dt, market="ALL")
    stock_price = stock_price.reset_index()
    stock_price.rename(columns={'티커':'종목코드'}, inplace=True)

    data = pd.merge(stock_data, stock_price, left_on='종목코드', right_on='종목코드', how='left')
    
    market_cap = stock.get_market_cap(dt)['시가총액']
    market_cap = pd.DataFrame(market_cap)

    data = pd.merge(data, market_cap, left_on='종목코드', right_on='티커', right_index=True, how='left')
    
    return data
```

```python
stock_data = get_sector('20220801')
```

## 재무정보 가져오기

```python
def get_data(code):
    #code = '005930'
    URL = f"https://finance.naver.com/item/main.nhn?code={code}"
    r = requests.get(URL)
    df = pd.read_html(r.text)[3]
    df = df.T
    df.columns = df.iloc[0]
    df.drop('주요재무정보', inplace=True)
    df.index = df.index.get_level_values(1)
    df.index.name = None
    df.drop(df.index[df.index.str.endswith('(E)')].tolist(), inplace=True)
    df.index = pd.to_datetime(df.index, format='%Y.%m')
    df.columns.name = '주요재무정보'
    
    return df
```

```python
def get_info(code):
    '''
    2018-2022까지의 연간 비율
    - 안정성비율 : '유동비율', '당좌비율', '부채비율', '유보율', '순차입금비율', '이자보상배율', '자기자본비율'
    - 성장성비율 : '매출액증가율', '판매비와관리비증가율', '영업이익증가율', 'EBITDA증가율', 'EPS증가율'
    - 수익성비율 : '매출총이익율', '세전계속사업이익률', '영업이익률', 'EBITDA마진율', 'ROA', 'ROE', 'ROIC'
    - 활동성비율 : '총자산회전율', '총부채회전율', '총자본회전율', '순운전자본회전율'
    '''
    #code='005930'
    URL = f"http://comp.fnguide.com/SVO2/ASP/SVD_FinanceRatio.asp?pGB=1&gicode=A{code}"
    r = requests.get(URL)
    tmp = pd.read_html(r.text)
    tmp = tmp[0]
    tmp.iloc[:,0] = tmp.iloc[:,0].apply(lambda x: x.replace('계산에 참여한 계정 펼치기',''))
    tmp = tmp.set_index(tmp.columns[0])
    tmp.index.name = None
    tmp = tmp.T
    
    return tmp
```

```python
def get_comparison(code, n):
    URL = f"https://comp.fnguide.com/SVO2/ASP/SVD_main.asp?gicode=A{code}"
    r = requests.get(URL)
    tmp = pd.read_html(r.text)[n].set_index('구분')
    
    return tmp.T
```
